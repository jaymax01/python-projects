# Analysing Customers Purchase Data

The Chinook music store's customers' information and purchase data is analysed using SQLite and some business questions are answered by passing queries to extract the relevant information for analysis. Business questions such as top selling genres in the USA, sales employee performance and sales by country are visualized to draw to make recomendations for the marketing team.

```python
# Creating functions to run queries, commands and display tables

import sqlite3 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline

def run_query(q):
    with sqlite3.connect('chinook.db') as conn:
        return pd.read_sql(q, conn)

def run_command(c):
    with sqlite3.connect('chinook.db') as conn:
        conn.isolation_level = None
        conn.execute(c)
    
def show_tables():
    q =  '''SELECT name, type
          FROM sqlite_master
          WHERE type IN ("table","view");'''
    return run_query(q)

show_tables()

```

```python
# Displaying the top selling genres in the US


q =  '''SELECT gn.name genre, 
       COUNT(t.track_id) tracks_sold, 
       CAST(COUNT(t.track_id) AS FLOAT) / 
       (SELECT COUNT(track_id) Total_sold FROM invoice_line il
       INNER JOIN invoice i ON i.invoice_id = il.invoice_id
       INNER JOIN customer c ON c.customer_id = i.customer_id
       WHERE c.country = "USA") percent_track_sold
       FROM genre gn 
       INNER JOIN track t ON t.genre_id = gn.genre_id
       INNER JOIN invoice_line il ON il.track_id = t.track_id
       INNER JOIN invoice i ON i.invoice_id = il.invoice_id
       INNER JOIN customer c ON c.customer_id = i.customer_id
       WHERE c.country = "USA"
       GROUP BY 1 
       ORDER BY 2 DESC
       LIMIT 10;'''

run_query(q)
```

```python
us_track_sales = run_query(q)
us_track_sales.plot.bar(x='genre',y='tracks_sold')
```

The store should also consider getting Rock artists from their record label since Rock accounts for more than half (53%) the total track sales in the US. Based on the above results of the top selling genres the following albums should be added to the store

* Red Tone (Punk)
* Slim Jim Bites (Blues)
* Meteor and the Girls (Pop)



```python
# Analysing sales support agent's performance

r = '''
SELECT e.employee_id, 
e.reports_to,
e.last_name || " " || e.first_name employee,
e.hire_date, 
e.country employee_country,
c.country customer_country,
SUM(i.total) total_employ_sales
FROM employee e
INNER JOIN customer c ON c.support_rep_id = e.employee_id
INNER JOIN invoice i ON i.customer_id = c.customer_id
GROUP BY 1
ORDER BY 4 DESC; '''

run_query(r)
```

```python
employee_sales = run_query(r)
employee_sales.plot.bar(x='employee', y='total_employ_sales',rot=45)
```

From the above results Canada has the top 3 sales support agents in the company. 
The difference in total sales for each of the agents especially between Jane and Steve can be explained by their hire dates and that her customers are in Canada as well.
Their perfomance could as well be attributed to the fact that they report to the same manager.


```python
# Analysing customer's sales data from different countries

d = '''
WITH cust_val AS
(SELECT
CASE
       WHEN (
            SELECT count(*)
            FROM customer
            where country = c.country
            ) = 1 THEN "Other"
           ELSE c.country
       END AS country,
       c.customer_id,
       il.*
FROM invoice_line il
INNER JOIN invoice i ON i.invoice_id = il.invoice_id
INNER JOIN customer c ON c.customer_id = i.customer_id)

SELECT

country,
total_no_customers,
total_val_sales,
av_val_sales_customer,
av_order_val

FROM
(SELECT
country,
COUNT(DISTINCT customer_id) total_no_customers,
SUM(unit_price) total_val_sales,
SUM(unit_price) / COUNT(DISTINCT customer_id) av_val_sales_customer,
SUM(unit_price) / COUNT(DISTINCT invoice_id) av_order_val,
CASE
    WHEN country = "Other" THEN 1
    ELSE 0
    END 
FROM cust_val
GROUP BY country
ORDER BY 6, 3 DESC
);

'''

run_query(d)
```

```python
country_ind = run_query(d)
country_ind.set_index("country", drop=True, inplace=True)
colors = [plt.cm.Accent(i) for i in np.linspace(0, 1, country_ind.shape[0])]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 11))
ax1, ax2, ax3, ax4 = axes.flatten()
fig.subplots_adjust(hspace=.5, wspace=.3)


# Top left pie plot
country_sales = country_ind["total_val_sales"].copy().rename('')
country_sales.plot.pie(
    ax=ax1,
    startangle=-90,
    counterclock=False,
    title='Sales by Country,\nNumber of Customers',
    colormap=plt.cm.Accent,
    fontsize=8,
    wedgeprops={'linewidth':0})


# Top right bar plot
cols = ["total_no_customers","total_val_sales"]
cust_vs_sales = country_ind[cols].copy()
cust_vs_sales.index.name = ''
for c in cols:
    cust_vs_sales[c] /= cust_vs_sales[c].sum() / 100 
cust_vs_sales.plot.bar(
    ax=ax2,
    colormap=plt.cm.Set1,
    title="Percent Customers vs Sales")
ax2.tick_params(top="off", right="off", left="off", bottom="off")
ax2.spines["top"].set_visible(False)
ax2.spines["right"].set_visible(False)

# Bottom left bar plot
avg_order = country_ind["av_order_val"].copy()
avg_order.index.name = ''
diff_from_avg = avg_order * 100 / avg_order.mean() - 100
diff_from_avg.drop("Other", inplace=True)
diff_from_avg.plot.bar(
    ax=ax3,
    color=colors,
    title="Average Order,\nPercentage Difference from Mean")
ax3.tick_params(top="off", right="off", left="off", bottom="off")
ax3.axhline(0, color='k')
ax3.spines["top"].set_visible(False)
ax3.spines["right"].set_visible(False)
ax3.spines["bottom"].set_visible(False)


# Bottom right bar plot
spc = country_ind["av_val_sales_customer"].copy()
spc.index.name = ''
spc.drop("Other",inplace=True)
spc.plot.bar(
    ax=ax4,
    color=colors,
    title="Average value of sales per customer, Dollars")
ax4.tick_params(top="off", right="off", left="off", bottom="off")
ax4.spines["top"].set_visible(False)
ax4.spines["right"].set_visible(False)

plt.show()
```

The US, Canada and Brazil are the top countries for customer sales for the company with total number of customers about matching the number of sales which indicates those countries have an established market for the company's product. 

Judging from the difference from the mean and the average value of sales per customer the countries with a potential for growth are The Czech Republic, The United Kingdom, India and Germany. More data however has to be collected to make better recommendations to the marketing team after test runs of marketing campaigns for new customers have been completed over a specified period.   

```python
# Catigorizing each invoice as an album or not

r = '''
CASE
    WHEN 
    (SELECT invoice_id, track_id
    FROM invoice_line il 
    EXTRACT
    SELECT t.track_id, a.album_id
    FROM track t
    INNER JOIN album a ON a.album_id = t.album_id) IS NULL THEN "YES"
    ELSE "NO"
END AS invoice_as_album

'''
run_query(r)

```
