# Finding the Best Markets to advertise E-learning products

A company that offers courses on programming wants to market its web and mobile development courses as well as its other coursework in data science, game development and other techincal coursework on its platform. 

The project aims to find the two best markets that the company could advertise their product in. Data from a platform that offers similar products will be used for the analysis. 

# Understanding the data 

Conducting surveys for a couple of different markets to find the ones that are best for advertising is thebest choice for data collection however it takes time and is expensive. An alternative is collecting exsisting data relevant to our analysis. 

A good source is from the free e-learning platform freeCodeCamp. The platform attracts new coders with different interest from its popular medium publication (over 400000 followers).  

The survey collected information about users on the platform and how they learned how to code. The link to the survey is publicly available on GitHub [here](https://github.com/freeCodeCamp/2017-new-coder-survey)

```python
# Reading in the datafile and exploring the data 

import pandas as pd 
survey_data = pd.read_csv('2017-fCC-New-Coders-Survey-Data.csv', low_memory = 0)
pd.options.display.max_columns = 150
print(survey_data.shape)
survey_data.head(5)
```

The columns in the data set are explanitory. The initial survey questions can be used to understand the column names and they can be found [here](https://github.com/freeCodeCamp/2017-new-coder-survey). Important columns for the analysis include CountryLive, EmploymentField, JobRoleInterest, JobPref, AttendedBootcamp, SchoolMajor, and MoneyForLearning

# Veryfing sample representivity

The population of interest for the analysis is the number of new coders that are interested in the programs offered by the company. The data that is available is considered as sample data. The JobRoleInterest column from the survey has information about the roles each participant is interested in working in. The following information about new coders will also be relevant to the sample data.

* Where the new coders are located
* What are the locations with the greatest number of new coders
* How much money are new coders willing to spend on learning.

A value count of respondents subject areas of interest especially the ones offered by the company and their frequencies will provide insights into how representative the data is of the population.  

```python
# Generating percentage frequenies for the JobRoleInterest column

survey_data['JobRoleInterest'].value_counts(normalize = True)*100
```

From the frequency table the majority of repondents are interested in Full-Stack Web Developering and other Web developing programs. Few people are interested in mobile development and other areas other than the latter. They are also interested in more than one subject as well. A count of the number of respondents for each area of interest is done below. 

```python
# Removing null values and counting areas of interests for respondents

interests_no_null = survey_data['JobRoleInterest'].dropna()
interest_areas_all = interests_no_null.str.split(',')
no_interest_areas = interest_areas_all.apply(lambda x: len(x))
no_interest_areas.value_counts(normalize = True).sort_index() * 100
```

31.7% of the respondents have a single subject area of interest while the others have more than one subject area of interest. To the benefit of the company programs are offered in areas other than mobile and web development making the survey data representative of the population data. 

The focus of the analysis is web and mobile development. The plot below will depict how many of the respondents are interested in both subject areas.

```python
web_mobile_counts = interests_no_null.str.contains('Web Developer|Mobile Developer')
freq_table = web_mobile_counts.value_counts(normalize = True)*100
print(freq_table)

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
freq_table.plot.bar(title = 'Respondents interested in at least one  subject \n Web and Mobile Developer')
plt.ylim([0,100])
plt.ylabel('Percentage')
plt.xticks([0,1],['Web or mobile \ndevelopment', 'Other subjects'], rotation = 0)

```

From the bar plots the majority of the repondents 86% are interested in the web and mobile development the area of the analyis. With the information available the following needs to be determined to determine the markets to advertise the company's product.

* Where are these new coders located.
* What are the locations with the greatest number of new coders.
* How much money new coders are willing to spend on learning.

# Finding the location and density of new coders

To find where the new coders live the CountryLive column describes the country where the respondent recides. Analysing this information provides insight into which countries the company can run their ad campaign. 

Countries will be considered as markets and an indicator of a good market is the country's number of potential customers. A frequency table of the CountryLive variable is presented below.  

```python
# Generating a frequency table for the CountryLive column

# Removing null values from the data based on job role interest
survey_data_good = survey_data[survey_data['JobRoleInterest'].notnull()].copy()

country_freq = survey_data_good['CountryLive'].value_counts()
country_rel_freq = survey_data_good['CountryLive'].value_counts(normalize = True)*100

freq_dict = {'Absolute Frequency': country_freq, 'Percentage': country_rel_freq}
df_freq_dict = pd.DataFrame(freq_dict)
df_freq_dict
```

The frequency table shows the United States as the most attractive market to advertise the company's products with 45.7% of all repondents. The second India with 7.7% and the United Kingdom with 4.6%. As much as this information is good for ad campaign purposes it is important to know how much respondents are willing to spend on learning in these markets.

# Finding the money spent on learning

The next step in the analysis is to find the amount of money the average respondent is willing to spend on learning in each country because markets that predominantly have free learning materials for new coders will not be profitable for the company. 

The MoneyForLearning column describes how much money the respondent spends on training. The company sells sunscriptions for $59 per month so it will be ideal to find out how much each student spends on average for training per month. The analysis will be restricted to the top four countries that has English as an official language which are The United States, India, The United Kingdom and Canada. 

```python
# Removing rows with null values and finding spending in each market

survey_data_good['MonthsProgramming'].replace(0,1, inplace = True)
survey_data_good['AmountSpentPerMonth'] = survey_data_good['MoneyForLearning'] / survey_data_good['MonthsProgramming']
survey_data_good['AmountSpentPerMonth'].isnull().value_counts()
```

There are 675 rows with null values they will be removed to continue the analysis

```python
sur_data_good = survey_data_good[survey_data_good['AmountSpentPerMonth'].notnull()]
sur_data_good['AmountSpentPerMonth'].isnull().value_counts()
```

The remaining data now needs to be grouped by the country markets where the company plans to run its adds in order to find the mean amount spent by a student in each country on learning. 

```python
# Removing rows with null values in the CountryLive column

sur_data_good = sur_data_good[sur_data_good['CountryLive'].notnull()]
sur_data_good['CountryLive'].isnull().value_counts()
sur_data_good['CountryLive'].value_counts().head()
```

```python
# Grouping the remaining data by the CountryLive column 

country_mean = sur_data_good.groupby('CountryLive').mean()

# Finding the mean amount spent by the top four countries

country_mean['AmountSpentPerMonth'][['United States of America',
                            'India', 'United Kingdom',
                            'Canada']]
```

The surprising result from the analysis is that on average Indian students spends more money on learning compared to The United Kingdom and Canada considering the net GDP of these countries compared to India. Peharps there is not enough representative data from The UK and Canada and also peharps respondents from India for the survey on average where those who spent more on learning compared their UK and Canadian counterparts. 

The next thing to investigatу are outliers in the data that could explain the above results. 

# Visualizing extreme outliers

The distribution of the US, India, the UK and Canada's spending per respondent are depicted by the following box plots

```python
## Removing the four countries of interest from the dataset
four_list = ['United States of America', 'India', 'United Kingdom','Canada']
four_int = sur_data_good[sur_data_good['CountryLive'].isin(four_list)]
# four_int.CountryLive.value_counts()

import seaborn as sns
sns.boxplot(y = 'AmountSpentPerMonth', x = 'CountryLive', data = four_int)
plt.title('Money Spent Per Month Per Country\n(Distributions)',
         fontsize = 16)
plt.ylabel('Money per month (US dollars)')
plt.xlabel('Country')
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada']) 
plt.show()
```

India, Canada and the UK have outliers but not extreme. The maximum average monthly amount spent in the three countries is \$10000 compared to the United States that has as much as $80000 a month. Up to \$20000 seems reasonable for most students to spend every month so spending over \$20000 is removed from the dataset.  

```python
# Recomputing the mean of the data without outliers
data_no_outl = survey_data_good[survey_data_good['AmountSpentPerMonth'] < 20000]
countries_mean = data_no_outl.groupby('CountryLive').mean()
countries_mean['AmountSpentPerMonth'][['United States of America',
                            'India', 'United Kingdom',
                            'Canada']]
```

```python
# Visualizing the data again with boxplots
four_only = data_no_outl[data_no_outl['CountryLive'].isin(four_list)]

sns.boxplot(y = 'AmountSpentPerMonth', x = 'CountryLive',
            data = four_only)
plt.title('Money Spent Per Month Per Country\n(Distributions)',
         fontsize = 16)
plt.ylabel('Money per month (US dollars)')
plt.xlabel('Country')
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada']) 
plt.show()
```

For India there are a few extreme outliers with spending over \$2500 per month. There are two of such people in India. Exploring the other columns of the dataset such as if they attended a bootcamp could give insights as to why the spending is justified. 

```python
# Removing India's outliers

india_outliers = four_only[(four_only['CountryLive'] == 'India') & (four_only['AmountSpentPerMonth'] >= 2500)]
india_outliers
```

All of the respondents that spent more than \$2500 in India did not attend a coding bootcamp. It is unclear where they got their extra code training and if that explains their extra spending apart from perhaps the survey question "Aside from university tuition, about how much money have you spent on learning to code so far (in US dollars)?" which made the respondents to include their university tuition in their expenses. These outliers can be removed from the dataset.
