# An analysis of the Star Wars movie series

Before the release of the Start Wars movie *The Force Awakens *was released a survey was taken to determine whether the movie in the Star Wars series *The Empire Strikes back *to determine if it is the best in the series. The responses of star wars fans where collected online using SurveyMonkey. The file can be found  [here](https://github.com/fivethirtyeight/data/tree/master/star-wars-survey).
The following  are the most important columns of information that were collected from the survey
* RespondentID - An anonymized ID for the respondent (person taking the survey)
* Gender - The respondent's gender
* Age - The respondent's age
* Household Income - The respondent's income
* Education - The respondent's education level
* Location (Census Region) - The respondent's location
* Have you seen any of the 6 films in the Star Wars franchise? -   Has a Yes or No response
* Do you consider yourself to be a fan of the Star Wars film       franchise? - Has a Yes or No

```python
# Reading and exploring the data

import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars.head(10)
```

```python
star_wars.columns
```

```python
star_wars_IDs = pd.notnull(star_wars['RespondentID'])
star_wars_IDs
```

```python
star_wars = star_wars[star_wars_IDs]
star_wars.head()
```

```python
# converting the Yes/No column responses to boolean type

dict_map = {'Yes':True, 'NO':False}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(dict_map)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(dict_map)
star_wars.head()
```

```python
# converting star wars movies seen to boolean values


import numpy as np
columns = star_wars.columns[3:9]
movie_map = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in columns:
    star_wars[col] = star_wars[col].map(movie_map)
```

```python
columns = {"Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"}

star_wars = star_wars.rename(columns=columns)
star_wars.head()

    
```

```python
# Converting columns ranking the movie series to float type
# Renaming the columns

star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
col_rename = {"Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"}
star_wars = star_wars.rename(columns = col_rename)
star_wars.head()
```

```python
star_wars.columns
```

```python
# Finding the higest ranked movie

star_wars_rancols = star_wars[star_wars.columns[9:15]]
star_wars_rancols.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
x = 6
fig, ax = plt.subplots()
plt.bar(range(x), star_wars_rancols.mean())
plt.xticks(rotation=60)
ax.set_xticks(range(x))
ax.set_xticklabels(['ranking_1', 'ranking_2', 'ranking_3', 'ranking_4', 'ranking_5',
       'ranking_6'])
```

So far the dataset has been cleaned, columns have been renamed and the data type for the ranking columns changed. From the plot the earlier film releases ranked higher because fans rate the other films in the series based on the earlier releases. 

```python
# Finding the number of people that have seen each movie

star_wars_seen = star_wars[star_wars.columns[3:9]].sum()
print(star_wars_seen)
```

```python
star_wars_seen.plot.bar()
```

From the plot the most seen movies were the original releases which also had high rankings by people who had gone to see the movies.   

```python
# Analysing movie ranking by gender

males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]

x = 6
fig, ax = plt.subplots()
plt.bar(range(6), males[males.columns[9:15]].mean())
plt.xticks(rotation=60)
ax.set_xticks(range(x))
ax.set_xticklabels(['ranking_1', 'ranking_2', 'ranking_3', 'ranking_4', 'ranking_5',
       'ranking_6'])
plt.show()

x = 6
fig, ax = plt.subplots()
plt.bar(range(6), females[females.columns[9:15]].mean())
plt.xticks(rotation=60)
ax.set_xticks(range(x))
ax.set_xticklabels(['ranking_1', 'ranking_2', 'ranking_3', 'ranking_4', 'ranking_5',
       'ranking_6'])
plt.show()

```

```python

males[males.columns[3:9]].sum().plot.bar()
plt.show()

females[females.columns[3:9]].sum().plot.bar()
plt.show()
```

Both men and women had high ratings for the original series releases however, more men than women went to see series 1 to 3 and women liked series 4, 5 and 6 more than men.
