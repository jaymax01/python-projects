# Analysing Employee Exit Survey Data 

The project involved analysing survey data done by the Department of Education, Training and Employment (DETE) and the Technical and Further Education institute (TAFE) in Queensland Australia. The links to the survey results for TAFE can be found [here](https://data.gov.au/dataset/ds-qld-89970a3b-182b-41ea-aea2-6f9f17b5907e/details?q=exit%20survey) and for DETE [here](https://data.gov.au/dataset/ds-qld-fe96ff30-d157-4a81-851d-215f2a0fe26d/details?q=exit%20survey) 
The stakeholders wanted to have answers to the following questions from the analysis
* Are employees who only worked for the institutes for a short period of time resigning due to some kind of dissatisfaction? What about employees who have been there longer?

The columns used for the analysis from the DETE and TAFE datasets are ID, SeparationType, Date, DETE Start Date, Record ID, Reason for ceasing employment, Overall Length of Service at Institute 



```python
# Reading both DETE and TAFE survey datasets

import pandas as pd
import numpy as np
dete_survey = pd.read_csv('dete_survey.csv')
tafe_survey = pd.read_csv('tafe_survey.csv')
```

```python
# Exploring the DETE dataset

dete_survey.info()
dete_survey.head()
```

```python
# Exploring the TAFE dataset

tafe_survey.info()
tafe_survey.head()
```

```python
# Counting the number of null values for the columns in both datasets

print(dete_survey.isnull().sum())
print()
print(tafe_survey.isnull().sum())
```

From the data summary about the both dataframes both have columns that have the same information and can be renamed as well as columns with information that is irrelivant to the analysis. For the DETE dataset most of the relivant columns have non null values while the other columns have missing or NAN values. The TAFE dataset has a lot of NAN or missing values and should be used to suppliment the DETE dataset.  

```python
# Reading the DETE dataset again replacing NAN values with Not Stated
dete_survey = pd.read_csv('dete_survey.csv', na_values='Not Stated')

# Identifing and removing irrelivant columns 
dete_survey_updated = dete_survey.drop(dete_survey.columns[28:49], axis=1)
tafe_survey_updated = tafe_survey.drop(tafe_survey.columns[17:66], axis=1)

# Columns left in both datasets
print(dete_survey_updated.columns)
print(tafe_survey_updated.columns)
```

NAN values in the DETE dataset where renamed Not Stated when reading the dataset again to capture the total number of missing values in the dataset not just NAN values. Redundant columns were removed to make the analysis easier to answer the stakeholders' questions

```python
# Renaming the columns for the DETE dataset

dete_survey_updated.columns = dete_survey_updated.columns.str.lower().str.strip().str.replace(' ', '_')
print(dete_survey_updated.columns)
```

```python
# Updating the column names of the TAFE dataset

col_name_mapping = {'Record ID': 'id', 'CESSATION YEAR': 'cease_date', 'Reason for ceasing employment': 'separationtype', 'Gender. What is your Gender?': 'gender', 'CurrentAge. Current Age': 'age',
       'Employment Type. Employment Type': 'employment_status',
       'Classification. Classification': 'position',
       'LengthofServiceOverall. Overall Length of Service at Institute (in years)': 'institute_service',
       'LengthofServiceCurrent. Length of Service at current workplace (in years)': 'role_service'}
tafe_survey_updated = tafe_survey_updated.rename(col_name_mapping, axis = 1)
tafe_survey_updated.columns
```

```python
dete_survey_updated.head()
```

```python
tafe_survey_updated.head()
```

Changes to the column names for both DETE and TAFE datasets where done to make column names more readable in the standard format and to make combining columns with the same information easier. 

```python
# Filtering the DETE and TAFE datasets

# Finding unique values for the seperation type columns
dete_survey_updated['separationtype'].value_counts()
```

```python
tafe_survey_updated['separationtype'].value_counts()
```

```python
# Choosing resignation for the separationtype columns
dete_survey_updated['separationtype'] = dete_survey_updated['separationtype'].str.split('-').str[0]
dete_resignations = dete_survey_updated[dete_survey_updated['separationtype'] == 'Resignation'].copy()
tafe_resignations = tafe_survey_updated[tafe_survey_updated['separationtype'] == 'Resignation'].copy()
```

```python
dete_survey_updated['separationtype'].value_counts()
```

The DETE dataset three resignation seperation types to make filtering of both datasets without errors only the word Resignation was retained for every entry of the separationtype column for the DETE dataset. 

```python
# Verifying and removing errors from the data

# Verifying cease dates and start date columns
dete_resignations['cease_date'].value_counts()
```

```python
dete_resignations['cease_date'] = dete_resignations['cease_date'].str.split('/').str[-1].astype('float')
dete_resignations['cease_date'].value_counts()
```

```python
dete_resignations['dete_start_date'].value_counts().sort_values()
```

```python
tafe_resignations['cease_date'].value_counts().sort_values()
```

The cease dates in the DETE and TAFE datasets do not allign for instance the year 2009 in the TAFE dataset does not have any corresponding data in the TAFE dataset. Since the analysis is not nessesarily focused on resignations by year the cease and start dates will be left the way they are.

```python
# Creating a new column for the DETE dataset for storing years of service

dete_resignations['institute_service'] = dete_resignations['cease_date'] - dete_resignations['dete_start_date']
dete_resignations['institute_service'].head()
```

To know the reason for resignations for employees for a short or longer period of time due to some dissatifaction the DETE dataset needed a column to store service years data. The institute_service column was created by subtracting the cease_date column from the dete_start_date column. 

```python
# Identifying The Dissatified Employees

tafe_resignations['Contributing Factors. Dissatisfaction'].value_counts()
```

```python
tafe_resignations['Contributing Factors. Job Dissatisfaction'].value_counts()
```

```python
def update_vals(val):
    if pd.isnull(val):
        return np.nan
    elif val == '-':
        return False
    else:
        return True
tafe_resignations['dissatisfied'] = tafe_resignations[['Contributing Factors. Job Dissatisfaction', 'Contributing Factors. Dissatisfaction']].applymap(update_vals).any(axis =1, skipna=False)
tafe_resignations_up = tafe_resignations.copy()
tafe_resignations_up['dissatisfied'].value_counts(dropna=False)
```

```python
dete_resignations['dissatisfied'] = dete_resignations[['job_dissatisfaction',
       'dissatisfaction_with_the_department', 'physical_work_environment',
       'lack_of_recognition', 'lack_of_job_security', 'work_location',
       'employment_conditions', 'work_life_balance',
       'workload']].applymap(update_vals).any(axis=1, skipna=False)
dete_resignations_up = dete_resignations.copy()
dete_resignations_up['dissatisfied'].value_counts(dropna=False)
```

A dissatified column was added to both DETE and TAFE datasets to determine if an employee was dissatified and resigned based on the columns in both datasets that contribute to dissatifaction. The values for the columns were changed to TRUE if the employee resigned because of dissatifaction, FALSE if not and NAN is no value was available.  

```python
# Combining both datasets

dete_resignations_up['institute'] = 'DETE'
tafe_resignations_up['institute'] = 'TAFE'
combined = pd.concat([dete_resignations_up, tafe_resignations_up], ignore_index=True)
combined_updated = combined.dropna(thresh=500, axis=1).copy()
```

An institute column was added to both datasets to indentify the data from both datasets when they are combined. Columns with less than 500 non null values were removed since they will not enough information for the analysis.

```python
# Editing the institute_service column

combined_updated['institute_service'].value_counts(dropna=False)
```

```python
combined_updated['institute_service_up'] = combined_updated['institute_service'].astype('str').str.extract(r'(\d+)').astype('float')
combined_updated['institute_service_up'].value_counts(dropna=False)
```

```python
def year_to_stage(val):
    if val < 3:
        return 'New'
    elif 3 <= val <=6:
        return 'Experienced'
    elif 7 <= val <=10:
        return 'Established'
    elif val >= 11:
        return 'Veteran'
    else:
        return np.nan
    
combined_updated['service_cat'] = combined_updated['institute_service_up'].apply(year_to_stage)
combined_updated['service_cat'].value_counts()
```

The institute_service column in the combined dataset was used to extract service years in decimal format and assigned to a new column institute_service_up to convert the use service years into categories mentioned in [this article](https://www.businesswire.com/news/home/20171108006002/en/Age-Number-Engage-Employees-Career-Stage) .A function was used to assign the correct service year category and they were assigned to a new column in the combined dataset called service_cat.
